# Dataset Definition and Understanding

This notebook outlines the dataset structure for credit risk modeling, focusing on the three pillars of Basel II/III: PD, LGD, and EAD.

### 1. Structure of the Dataset
The dataset typically consists of historical loan data containing:
* **Static Information**: Borrower demographics (age, income, employment status).
* **Loan Attributes**: Loan amount, interest rate, term, purpose, and collateral type.
* **Behavioral Data**: Payment history, delinquency status, and credit utilization.
* **Economic Indicators**: Inflation rates, unemployment rates, or house price indices.

### 2. Variable Definitions

| Variable Category | Description |
| :--- | :--- |
| **Demographics** | Features like `age`, `income`, `home_ownership`. |
| **Credit History** | `credit_score`, `years_of_history`, `past_due_incidents`. |
| **Loan Specs** | `loan_amnt`, `int_rate`, `term_months`. |
| **Macroeconomic** | `gdp_growth`, `unemployment_index`. |

### 3. Target Variable Identification

To build credit risk models, we define three specific targets:

1.  **PD (Probability of Default)**:
    * **Target**: `is_default` (Binary: 1 if the borrower defaulted within a specific window, 0 otherwise).
2.  **LGD (Loss Given Default)**:
    * **Target**: `recovery_rate` or `loss_percentage` (The proportion of the exposure that cannot be recovered after default).
3.  **EAD (Exposure at Default)**:
    * **Target**: `total_exposure` (The total value a bank is exposed to at the time of default).
